### Flow Control


#### If conditions


```python
names <- c('John', 'Leo', 'Ahmed')
names_copy <- names                   # change in names_copy will not affect names

if ('John' %in% names & TRUE) {
  print('John exists')
} else if ('Leo' %in% names | FALSE) { 
  print('Leo exists')
} else { 
  print('Neither John nor Leo exists')
}
```

#### For loops


```python
for (name in names)
{
  print(name)
  if (name == 'Leo') {
    break                            # next to skip
  }
}
```

#### Plotting Points


```python
options(repr.plot.width=6, repr.plot.height=6)  
par(bg="white")                             # set background color of plots to white  (this has many plot options)            
x = 1:5
y = c(2, 4, 6, 8, 10)
plot(x,y, type="b")                           # type:  l=line, p=points, b=both, h for stems

# To add a plot on top:
#lines(x, y-0.3, col="red", type="b")       
```

```python
options(repr.plot.width=16, repr.plot.height=6)                           # don't let ituse the default (try commenting to see)
par(mfrow=c(1,2), bg='white')                                           # 1 row, 2 columns, white background
plot(x, type="b", col="red", lwd=2,  main="My Plot", xlab="X", ylab="Y")    # plots the given vector against itself
plot(x,sin(x), type="b", col="#4b84ff", lwd=2,  main="My Plot", xlab="X", ylab="Y")
```

## Vectors


Sequence of elements that share the same type


#### Types


```python
# There are six data types for vectors:
v1 <- c(1,2,3,4.5,2,3)                  # Numeric (double)
v2 <- c("tree","street","car")          # Character
v3 <- c(TRUE , FALSE ,TRUE)             # Logical
v4 <- c(23L, 192L, 0L)                  # Integer (we must use L to specify it is an integer)
v5 <- c(3+2i, 4-5i, -2i)                # Complex
v6 <- charToRaw("Hello")                # Raw (converts to ASCII)
printv(v6)

# Vectors can hold different types of data and in this case they are all cast to the most general of them
v7 <- c("data", 22.5, TRUE)
typeof(v7)
```

```python
# print the type of v1, v7
print(paste("v1 is of type", typeof(v1)))       # recall, paste is used to concatenate strings
print(paste("v7 is of type", typeof(v7)))
```

#### Extracting Elements


Positive index selects elements form specified positions


```python
v1 <- c(10, 20, 30, 40, 50, 60)
v1[1]                               # Indexing starts at 1
v1[c(2,4)]                          # Find the elements at positions 2 and 4
v1[c(4,4)]
v1[2:5]
v1[5:2]
v1[c(2.2,3.6)]                      # Real numbers are silently truncated to integers.
```

Negative index deselects elements form specified positions


```python
#Negative integers omit elements at the specified positions (this doesn't change the original vector):
v1 <- v1[-3]                        # Omit the element at position 3
v1[-c(2,1)]                         # Omit the elements at positions 2 and 1
```

#### Masking Vectors


```python
v1 <- c(1,2,3,4.5,3,2)
# Similar to masking in numpy
v1>2
v1==2
v1!=2
v1[v1>2]
```

## Matrix


Generalization of a vector, elements share the type as well.


```python
cells <- seq(10,90,by=10)       # We will give this flat vector to the matrix function. 
                                # What happens if we make 90 above into 80?
rnames <- c("R1", "R2","R3")    # Can optionally give row and column names
cnames <- c("C1", "C2","C3") 
mymatrix <- matrix(cells, nrow = 3, ncol = 3, byrow =TRUE, dimnames = list(rnames, cnames))
mymatrix
```

Access a specific element


```python
mymatrix[2,3]
```

Access a column as a vector


```python
mymatrix[,2]                # like mymatrix[:,2] in python
mymatrix[,"C2"]     
```

Access multiple columns as a matrix


```python
mymatrix[,c("C2","C3")]
```

Access a row as a vector


```python
mymatrix[1,]
```

Excluding elements


```python
mymatrix[-2,]       # Whole matrix except the second row
mymatrix[1,-3]      # first row and exclude the third column
```

## Dataframes


Dataframes are meant to represent real data. Each column is a vector of some type and each row is a data point.

This reminds you of lists? Quoting SO:
"
Data frames are lists as well, but they have a few restrictions:

you can't use the same name for two different variables

all elements of a data frame are vectors

all elements of a data frame have an equal length."


```python
# Let's make three columns
d <- c(1,2,3,4,4,4)
e <- c("red", "white", "red", NA,"red","red")
f <- c(TRUE,TRUE,TRUE,FALSE,FALSE,NA)
mydata <- data.frame(d,e,f)

# Let's give the columns names
colnames(mydata) <- c("ID","Color","Passed") 
mydata
```

Accessing a row or column


```python
mydata[1,]                              # extract first row of the data frame
mydata[2]                               # extract the second column (like mydata[,2] which is a vector)
```

Observe the differences


```python
mydata$ID               # the column ID
mydata[,1]
mydata[1]
mydata["ID"]
```

Accessing multiple columns


```python
mydata[1:2]                             # columns 2,3 of data frame
mydata[c("ID","Color")]                 # columns ID and color from data frame
```

Subsetting the data frame (like SQL)


```python
subdfm<- subset(mydata, ID <= 3, select=c(ID,Color))          # SELECT ID, COLOR FROM mydata WHERE ID <= 3
subdfm

subdfm<- subset(mydata, ID <= 3 & Color == 'red', select=c(ID,Color))
subdfm
```

An alternative method


```python
mydata[mydata$ID <= 3, c('ID', 'Color')]
#with(mydata, mydata[ID<=3, c('ID','Color')])
```
